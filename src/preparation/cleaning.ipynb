### Get Running Backs

Import data set:

```python
import pandas as pd

original = pd.read_csv("../../data/raw/Test2.csv")

original.head()
```

Fill in NaN values and find players with no position listed:

```python
data = original.copy()

# group data frame by each player id
for player_id, group in data.groupby(['playerid']):
    
    # get all of the listed positions of the player:
    positions = list(group['pos'])
    
    # if they were a running back at one point:
    if positions and 'RB' in set(positions):
        
        # fill all NaN as RB:
        for i in range(len(positions)):
            if positions[i] != positions[i]:
                positions[i] = 'RB'
        
        # update data:
        data.loc[data['playerid'] == player_id, 'pos'] = positions
```

You can see the updates:

```python
data.head()
```

Remove all non-RB's:

```python
data = data[data['pos'] == 'RB']
```

Write to file:

```python
data.to_csv('../../data/raw/listed_players.csv')
```

### Unlisted Players:

```python
unlisted_players = pd.DataFrame()

# group data frame by each player id
for player_id, group in original.groupby(['playerid']):

    # get all of the listed positions of the player:
    positions = list(group['pos'])
    
    # if the only position listed is nan
    if len(set(positions)) == 1 and (positions[0] != positions[0]):
        
        # add only first row of player to unlisted_players
        unlisted_players = unlisted_players.append(original[original['playerid'] == player_id].iloc[0])
```
