```python
import pandas as pd

frames = []
```

```python
for i in range(1997, 2021):
    df = pd.read_csv(f"../../data/raw/offensive stats/passing{i}.csv")
    df["Year"] = i
    frames.append()
```

```python
output = pd.concat(frames)
```

```python
output.assign(
    Team = lambda dataframe: dataframe['Team'].map(lambda team: team if team != "Football Team" else "Redskins") 
).reset_index()
```

```python
for col in output.columns:
    output[col] = output[col].map(lambda c: c.replace("T", "") if isinstance(c, str) else c ) # remove touchdown indicator
```
