```python
import pandas as pd

oline = pd.read_csv("../../data/raw/OLineData.csv")
data = pd.read_csv("../../data/processed/newdraft.csv")
```

```python
oline.head()
```

```python
oline = oline[["Year", "Team", "Adjusted Line yards", "Running Back Yards", "Stuffed Rate"]]
oline.columns = ["year", "team", "team_adjusted_line_yards", "team_running_back_yards", "team_stuffed_rate"]
```

```python
merged = data.merge(oline, on=["year", "team"], how="left")
```
