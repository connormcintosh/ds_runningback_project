```python
import pandas as pd

running_backs = pd.read_csv("../../data/processed/combined.csv")
receiving = pd.read_csv("../../data/raw/ReceivingStats.csv")
```

```python
receiving.head()
```

```python
receiving = receiving.rename(columns=
    {"yds": "yards", }
)
```

```python
total = pd.merge(running_backs.drop(columns=["g", "gs", "fumbles"]),
                 receiving.drop(columns=["playerid", "pos", "age"]),
                 on=['playername', 'year', 'team'],
                 suffixes=["_run", "_rec"])

total.to_csv("../../data/processed/total.csv")
```
